#### LaTeX Command Reference

**LaTeX Setup** (run this cell early)
$$
\newcommand{\cv}[1]{\begin{bmatrix} #1 \end{bmatrix}}
\newcommand{\mat}[1]{\begin{bmatrix} #1 \end{bmatrix}}
\newcommand{\norm}[1]{\left\| #1 \right\|}
\newcommand{\abs}[1]{\left| #1 \right|}
$$

Verify: $x = \cv{1 \\ 2}$

**Linear Algebra:**
- `\cv{1 & 2 & 3}`: $\cv{1 & 2 & 3}$
- `\cv{1 \\ 2 \\ 3}`: $\cv{1 \\ 2 \\ 3}$
- `\mat{1 & 2 \\ 3 & 4}`: $\mat{1 & 2 \\ 3 & 4}$
- `A^T`: $A^T$ 
- `A^{-1}`: $A^{-1}$ 
- `\norm{v}`: $\norm{v}$ 
- `\abs{x}`: $\abs{x}$ 
- `u \perp v`: $u \perp v$
- `\langle u, v \rangle`: $\langle u, v \rangle$
- `u \cdot v`: $u \cdot v$
- `u \times v`: $u \times v$
- `\det(A)`: $\det(A)$
- `\text{rank}(A)`: $\text{rank}(A)$
- `\text{trace}(A)`: $\text{trace}(A)$
- `\sum_{i=1}^{n} x_i`: $\sum_{i=1}^{n} x_i$

**Greek Letters:**
- `\lambda`: $\lambda$
- `\sigma`: $\sigma$
- `\Sigma`: $\Sigma$
- `\theta`: $\theta$
- `\alpha, \beta, \gamma`: $\alpha, \beta, \gamma$

**Special Sets & Symbols:**
- `\mathbb{R}`: $\mathbb{R}$
- `\mathbb{C}`: $\mathbb{C}$
- `\in`: $\in$
- `\subset`: $\subset$
- `\subseteq`: $\subseteq$
- `\emptyset`: $\emptyset$

**Subscripts & Superscripts:**
- `x_i`: $x_i$ 
- `x^2`: $x^2$ 
- `x_i^2`: $x_i^2$ 
- `x_{ij}`: $x_{ij}$ 

**Other Common:**
- `\vec{v}`: $\vec{v}$ 
- `\hat{x}`: $\hat{x}$ 
- `\bar{x}`: $\bar{x}$ 
- `\tilde{x}`: $\tilde{x}$ 
- `\dot{x}`: $\dot{x}$ 
- `\ldots`: $\ldots$
- `\infty`: $\infty$
- `\approx`: $\approx$
- `\leq, \geq`: $\leq, \geq$

**Arrows:**
- `\to` or `\rightarrow`: $\to$
- `\leftarrow` or `\gets`: $\leftarrow$
- `\leftrightarrow`: $\leftrightarrow$
- `\longrightarrow`: $\longrightarrow$
- `\longleftarrow`: $\longleftarrow$
- `\Rightarrow`: $\Rightarrow$
- `\Leftarrow`: $\Leftarrow$

- `\Longrightarrow`: $\Longrightarrow$

- `\mapsto`: $\mapsto$
