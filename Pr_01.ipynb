#### 4. Suppose A is the 3 by 3 matrix ones(3, 3) of all ones. Find two independent vectors x and y that solve Ax = 0 and Ay = 0. Write that first equation Ax = 0 (with numbers) as a combination of the columns of A. Why don't I ask for a third independent vector with Az = 0?

```julia
A=ones(3,3)
```

```julia
x1=[2,-1,-1]
```

```julia
A*x1
```

```julia
x2=[-1,2,-1]
```

```julia
A*x2
```

These are all of unique vectors that result in A*x = 0, but a1 == -a2 -a3, therefore, there are only two independent vectors. 

These solutions are in the NullSpace of A. Moreover, since there are two indendent vectors, the Rank of A is 2.

```julia
Z=stack([[2, -1, -1],
[-1, 2, -1],
[-1, -1, 2]])
```

```julia
rank(Z)
```

```julia
Z[1,:] == - Z[2,:] - Z[3,:] # Cols
```

```julia
Z[:,1] == - Z[:,2] - Z[:,3] # Rows
```
